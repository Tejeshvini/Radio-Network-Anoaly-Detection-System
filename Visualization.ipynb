# Date formatting on x axis

```python
import numpy as np
for bg_id in bg_id_pred:
    train_data_plot = train_data[train_data['bg_id'] == bg_id]
    prediction_data_plot = prediction_data[prediction_data['bg_id'] == bg_id]
    anomaly = prediction_data_plot.loc[prediction_data_plot['Predicted_vals'] >2,['Predicted_vals']]
    index_length = len(train_data_plot.index)
    print(index_length)
    step_value = int(index_length/10)
    ticks_to_use = train_data_plot.index[::step_value]
    labels = [i for i in ticks_to_use ]
    formatter = matplotlib.dates.DateFormatter('%H:%M:%S')
    plt.clf()
    figure, axes = plt.subplots() 
    #axes.xaxis_date()
    #t0=date2num(train_data_plot.index[0]) # x axis startpoint
    #t1=date2num(train_data_plot.index[-1])
    plt.title(bg_id)
    
    axes.plot(train_data_plot.index,train_data_plot["Time_delay"])
    axes.plot(prediction_data_plot.index, prediction_data_plot["Predicted_vals"],color='green')
    axes.scatter(anomaly.index,anomaly['Predicted_vals'], color='red')
    axes.set_xticks(ticks_to_use)
    axes.set_xticklabels(labels,rotation=45)
    figure.tight_layout()
    figure.savefig(str(bg_id)+'.png')
    plt.show()
```
