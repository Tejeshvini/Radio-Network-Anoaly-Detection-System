```python
combined_csv = pd.read_csv('all_pred_data_18.csv')
```

```python
predictions = []
for x in combined_csv['Predicted_vals']:
    y = x[1:-1]
    predictions.append(y)
```

```python
combined_csv['Predicted_vals'] = predictions
```

```python
bg_id = []
for x in combined_csv["bg_id"]:
    y = x[8:]
    bg_id.append(y)
```

```python
combined_csv["bg_id"] = bg_id
```

```python
combined_csv.head()
```

```python
combined_csv['Predicted_vals'] = combined_csv['Predicted_vals'].astype(str).astype(float)
combined_csv['bg_id'] = combined_csv['bg_id'].astype(str).astype(int)
```

```python
bg_id_pred = combined_csv["bg_id"].unique()
```

```python
bg_id_pred
```

```python
bg_id_timedelays = pd.DataFrame(columns=['BG_ID','Time delays > 2'])
```

```python
for bg_id in bg_id_pred:
    bg_id_data = combined_csv[combined_csv['bg_id'] == bg_id]
    anomaly = bg_id_data.loc[bg_id_data['Predicted_vals'] >2,['Predicted_vals']]
    print(bg_id)
    print(len(anomaly))
    data = [{'BG_ID':bg_id,'Time delays > 2':len(anomaly)}]
    print(data)
    bg_id_timedelays = bg_id_timedelays.append(data,ignore_index=True,sort=False)
    print(bg_id_timedelays)
```
