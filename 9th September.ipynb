Train and Test Split

```python
def test_train_split(current_bg_df):
    
    NumberOfElements = len(current_bg_df)
    #print(x, NumberOfElements)
    TrainingSize = int(NumberOfElements * 0.7)
    TrainingData = current_bg_df[0:TrainingSize]
    #TrainingData.to_csv(str(current_bg_df)+"_"+"train"+".csv")
    TestData = current_bg_df[TrainingSize:NumberOfElements]
    train_data = TrainingData['Time_delay']
    test_data = TestData['Time_delay']
    return(train_data,test_data)
```

```python
def create_csv(_test_data_,p,q,d,mse_,pred):
    column_names = ["Date_time","bg_id", "p", "q","d","Test_vals","Predicted_vals","mse"]
    prediction_ = pd.DataFrame(columns = column_names)
    prediction_['Date_time'] = _test_data_.index
    prediction_["bg_id"] = x
    prediction_["p"] = p
    prediction_["q"] = q
    prediction_["d"] = d
    prediction_["mse"] = mse_
    prediction_["Test_vals"] = _test_data_.values
    prediction_["Predicted_vals"] = pred
    prediction_ = prediction_.set_index('Date_time')
    prediction_.to_csv(str(x)+"_"+str(p)+"_"+str(q)+"_"+str(str(round(mse_,3)))+".csv")
```

Create subset dataframes

```python
dataframes = dict()
for bg_id in sample_bg_ids:
  dataframes['data_bg_{0}'.format(bg_id)] = data_new[data_new['Balise_group_ID'] == bg_id]
  a  = dataframes['data_bg_{0}'.format(bg_id)]
  b = a['Time_delay']
  dataframes['data_bg_{0}'.format(bg_id)] = b
```

# Main Loop

```python
for dataframe in dataframes:

    x = dataframe
    
    ##Making dfs based on locoid
    current_bg_df = pd.DataFrame(dataframes[x])
    #print(x)
    #print(current_bg_df)
    
    ##drop na values
    current_bg_df = current_bg_df.dropna()
    current_bg_df = current_bg_df.reset_index().drop_duplicates(subset='Date_Time', keep='last').set_index('Date_Time').sort_index()
    
    ##Test train split
    _train_data_, _test_data_ = test_train_split(current_bg_df)
    
    
    mse_1 = 100000
    print("Stating grid search for",x)
    for p in range(1,3):
        for q in range(1,3):
            try:
                stationarity = adfuller(_train_data_, autolag='AIC')
                p_value = stationarity[1]
                pred = []
                
                print(p_value)
                if p_value > 0.05:
                    d = 1
                else:
                    d = 0
    
                _train_data_1 = [y for y in _train_data_]
                
                for i in range(len(_test_data_)):
                    #print('Building model for',x,' with p and q as',p,q)
                    model_1 = ARIMA(_train_data_1, order=(p,d,q))
                    model_1fit = model_1.fit()
                    yhat = model_1fit.forecast()[0]
                    #print(yhat)
                    pred.append(yhat)
                    #print(_test_data_[i])
                    _train_data_1.append(_test_data_[i])
                    #print(_test_data_[i])
                    #print(i)
                mse_ = mean_squared_error(_test_data_,pred)
                print("MSE",x,mse_)
                if mse_ < mse_1 :
                    create_csv(_test_data_,p,q,d,mse_,pred)
                    mse_1 = mse_
            except:
                pass
```
